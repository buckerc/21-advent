```python
import numpy as np

inputs = []
outputs = []
temp   = []
sig_pattern = []
output_pattern = []
with open('./input.txt','r') as file:
    for line in file:
        ins, outs = line.split(' | ')
        sig_pattern = ins.strip().split(' ')
        output_pattern = outs.strip().split(' ')
        inputs.append(sig_pattern)
        outputs.append(output_pattern)
        
seven = 3
one = 2
eight = 7
four = 4
counter = 0
for outs in outputs:
    for o in outs:
        if len(o) in [one, four, seven, eight]:
            counter = counter + 1
counter
```

```python
convert = {}
convert[119] = 0
convert[36] = 1
convert[93] = 2
convert[109] = 3
convert[46] = 4
convert[107] = 5
convert[123] = 6
convert[37] = 7
convert[127] = 8
convert[111] = 9


digits = 0
for ins,outs in zip(inputs,outputs):

    maybe_three = []
    maybe_six = []
    for i in ins:
        if len(i)==2:
            one = i
        elif len(i) == 4:
            four = i
        elif len(i) == 3:
            seven = i
        elif len(i) == 7:
            eight = i
        elif len(i) == 5:
            maybe_three.append(i)
        elif len(i) == 6:
            maybe_six.append(i)

    for s in seven:
        if s in one:
            continue
        top = s

    middle = ''
    for f in four:
        if f in one:
            continue
        middle = middle + f

    for m in maybe_three[0]:
        if m in maybe_three[1] and m in maybe_three[2]:
            if m in middle:
                middle = m

    bottom = ''
    for m in maybe_three[0]:
        if m in maybe_three[1] and m in maybe_three[2]:            
            if m in [middle,top]:
                continue
            assert bottom == ''
            bottom = m

    six = ''
    for ms in maybe_six:
        mscount = 0
        for m in ms:
            if m in one:
                mscount = mscount + 1
        if mscount == 1:
            assert six == ''
            six = ms
            break

    bottom_right = ''
    for o in one:
        if o in six:
            assert bottom_right == ''
            bottom_right = o
            break

    top_right = ''        
    for o in one:
        if o == bottom_right:
            continue            
        assert top_right == ''
        top_right = o
        break

    for f in four:
        if f in [middle, top_right, bottom_right]:
            continue
        top_left = f
        
    bottom_left = ''
    for e in eight:
        if e in [top, top_left, top_right, middle, bottom_right, bottom]:
            continue
        bottom_left = e

    counter = 0        
    for idx,ot in enumerate(outs):
        counter = 0
        for o in ot:
            if o == top:
                counter = counter + 1
            elif o == top_left:
                counter = counter + 2
            elif o == top_right:
                counter = counter + 4
            elif o == middle:
                counter = counter + 8
            elif o == bottom_left:
                counter = counter + 16
            elif o == bottom_right:
                counter = counter + 32
            elif o == bottom:
                counter = counter + 64
        digits = digits + convert[counter] * 10**(3-idx)
digits
```
