```python

def breath_first(history, stack,board):
    fin_x, fin_y = board.shape
    fin_x -= 1
    fin_y -= 1    
    while len(stack) > 0:
        print(len(stack))
        valid_options = []
        for s in stack:
            x = s['x']
            y = s['y']
            score = s['score']

            for xx,yy in zip([x,  x+1,x,  x-1], 
                             [y+1,y,  y-1,y]):
                if xx < 0 or xx > fin_x or yy < 0 or yy > fin_y:
                    continue
                temp = score + board[xx,yy] 
                if temp > history[xx,yy]:
                    continue

                valid_options.append({
                    'x': xx,
                    'y': yy,
                    'score': temp
                })

        to_pop = []
        for idx,v in enumerate(valid_options):
            x = v['x']
            y = v['y']
            score = v['score']      
            if score < history[x,y]:
                history[x,y] = score
            else:
                to_pop.append(idx)

        if len(to_pop)>1: 
            for tp in to_pop[::-1]:
                _ = valid_options.pop(tp)

        stack = valid_options

    return history[fin_x,fin_y]
```

```python

board = inputs

fin_x, fin_y = board.shape
history = {}
for x in range(fin_x):
    for y in range(fin_y):
        history[x,y] = 9*fin_x*fin_y


stack = []
stack.append({'x': 0,
        'y':0,
        'score':0,
})
history = breath_first(history,stack,board)
history
```

```python
multiple = 5
board = inputs

fin_x, fin_y = board.shape
big_board = np.zeros((fin_x*multiple,fin_y*multiple)).astype(int)
history = {}

for x in range(fin_x):
    for y in range(fin_y):
        b = board[x,y]
        for i in range(multiple):
            for j in range(multiple):
                big_board[x+i*fin_x,y+j*fin_y] = b+i+j
                history[x+i*fin_x,y+j*fin_y] = {'best': 9*fin_x*fin_y*multiple*multiple, 'who':-1}

big_board = np.where(big_board>9,big_board-9,big_board)            

fin_x = (fin_x * multiple) - 1
fin_y = (fin_y * multiple) - 1

stack = []
stack.append({'x': 0,
        'y':0,
        'score':0,
        'id':0
})

history = breath_first(history,stack,big_board)
history
```
